```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import numpy as np  # useful for many scientific computing in Python
# use the inline backend to generate the plots within the browser
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')  # optional: for ggplot-like style

# check for latest version of Matplotlib
print('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
# your code goes here
QUERY = """
SELECT ConvertedComp,COUNT(*) as Count
FROM master
group by ConvertedComp
order by ConvertedComp
"""
df=pd.read_sql_query(QUERY,conn)
df.head()
```

```python
# your code goes here

df['ConvertedComp'].plot(kind='hist', figsize=(8, 5))


#df['ConvertedComp'].plot(kind='hist', figsize=(8, 5))

# add a title to the histogram
plt.title('Histogram of ConvertedComp')
# add y-label
plt.ylabel('Compensation in $')
# add x-label
plt.xlabel('ConvertedComp')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
df2 = pd.read_sql_query(QUERY,conn)
df2['Age'].plot(kind='box', figsize=(10, 7))

plt.title('Box plots of age')
plt.ylabel('Age in years')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age,WorkWeekHrs
FROM master
order by Age
"""
df3 = pd.read_sql_query(QUERY,conn)
df3.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10, 6), color='darkred')

# add title and label to axes
plt.title('scatter plot of Age and Work-week Hours')
plt.xlabel("Age (years)")
plt.ylabel("Work-week (Hours)")

# show plot
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import plotly.express as px
QUERY = """
SELECT WorkWeekHrs,CodeRevHrs,Age
FROM master
"""
df4 = pd.read_sql_query(QUERY,conn)
bub_data = df4.groupby(['WorkWeekHrs', 'CodeRevHrs'])['Age'].sum().reset_index()
bub_data.head()
```

```python
# your code goes here
# Create bubble chart here
fig = px.scatter(bub_data, x="WorkWeekHrs", y="CodeRevHrs", size="Age",
                 hover_name="WorkWeekHrs", title='Work-week hours vs Code review hours', size_max=60)
fig.show()
```

Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.

```python
# your code goes here
colors_list = ['yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']

df5['count'].plot(kind='pie',figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            colors=colors_list  # add custom colors
                            )

# scale the title up by 12% to match pctdistance
plt.title('Top 5 databases that respondents wish to learn next year', y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=df5['DatabaseDesireNextYear'], loc='upper left') 

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs,CodeRevHrs,Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df6 = pd.read_sql_query(QUERY,conn)

print(df6.head())
print('Work-week hours median:', df6['WorkWeekHrs'].median())
print('Code Review hours median:', df6['CodeRevHrs'].median())
```

```python
# your code goes here
df7 = df6.groupby('Age').median()
df7.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median workweek hours and Code Review hours in 30-35 age group')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp,Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df8 = pd.read_sql_query(QUERY,conn)

print(df8.head())
print('ConvertedComp median:', df8['ConvertedComp'].median())
```

```python
#Graded Quiz: Visualizing Comparison of Data
# your code goes here
QUERY = """
SELECT ConvertedComp,Age
FROM master
WHERE Age BETWEEN 25 AND 30
"""
df16 = pd.read_sql_query(QUERY,conn)

print(df16.head())
print('ConvertedComp median:', df16['ConvertedComp'].median())
df17 = df16.groupby('Age').median()
df17.plot(kind='line', figsize=(20, 6))
plt.title('median ConvertedComp for all ages from 25 to 30 age group')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.show()
```
